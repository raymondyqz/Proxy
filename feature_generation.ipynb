```python
'''Gender Aggregation

Here we will count the number of
1. Founders
2. Male founders
3. Female founders

and the percentage of founders that are male.

'''
temp = df_All_Companies_merged['gender_of_founders'].values[:].tolist()

df_All_Companies_merged['number_of_founders'] = float('nan')
df_All_Companies_merged['number_of_male_founders'] = float('nan')
df_All_Companies_merged['number_of_female_founders'] = float('nan')
df_All_Companies_merged['percentage_of_male_founders'] = float('nan')

for idx in range(len(temp)):
    if not isinstance(temp[idx], str):
        if math.isnan(temp[idx]):
            continue
    temp2 = temp[idx].count(',') + 1
    df_All_Companies_merged.at[idx, 'number_of_founders'] = temp2
    df_All_Companies_merged.at[idx, 'number_of_male_founders'] = temp2 - temp[idx].count('female')
    df_All_Companies_merged.at[idx, 'number_of_female_founders'] = temp[idx].count('female')
    df_All_Companies_merged.at[idx, 'percentage_of_male_founders'] = (temp2 - temp[idx].count('female')) / (temp[idx].count(',') + 1)
```

```python
'''
USA vs the world
1. flag if company is in san francisco 
2. flag if founders are in san francisco 
3. flag is founders went to university in USA
4. flag for other major cities- New York, Miami
'''
temp1 = df_All_Companies_merged['city'].values[:].tolist()
temp2 = df_All_Companies_merged['city_of_founders'].values[:].tolist()

df_All_Companies_merged['in_sf'] = float('nan')
df_All_Companies_merged['in_ny'] = float('nan')
df_All_Companies_merged['in_miami'] = float('nan')
df_All_Companies_merged.at[idx, 'percentage_founders_in_sf'] = float('nan')
df_All_Companies_merged.at[idx, 'percentage_founders_in_ny'] = float('nan')
df_All_Companies_merged.at[idx, 'percentage_founders_in_miami'] = float('nan')


for idx in range(len(temp1)):
    if temp1[idx] == 'San Francisco':
        df_All_Companies_merged.at[idx, 'in_sf'] = 1
    else:
        df_All_Companies_merged.at[idx, 'in_sf'] = 0

for idx in range(len(temp1)):
    if temp1[idx] == 'New York':
        df_All_Companies_merged.at[idx, 'in_ny'] = 1
    else:
        df_All_Companies_merged.at[idx, 'in_ny'] = 0
        
for idx in range(len(temp1)):
    if temp1[idx] == 'Miami':
        df_All_Companies_merged.at[idx, 'in_miami'] = 1
    else:
        df_All_Companies_merged.at[idx, 'in_miami'] = 0
        
for idx in range(len(temp2)):
    if not isinstance(temp2[idx], str):
        if math.isnan(temp2[idx]):
            continue
    cities = temp2[idx].count(',') + 1
    sf = 0
    ny = 0
    miami = 0
    city_list = temp2[idx].split(',')
    for city in city_list:
        if city == 'San Francisco':
            sf += 1
        if city == 'New York':
            ny += 1
        if city == 'Miami':
            miami += 1
    df_All_Companies_merged.at[idx, 'percentage_founders_in_sf'] = sf/cities
    df_All_Companies_merged.at[idx, 'percentage_founders_in_ny'] = ny/cities
    df_All_Companies_merged.at[idx, 'percentage_founders_in_miami'] = miami/cities
        
df_All_Companies_merged[['city','city_of_founders','in_sf','in_ny','in_miami', 'percentage_founders_in_sf', 
                         'percentage_founders_in_ny', 'percentage_founders_in_miami']]

```

```python
# using string simularity to estimate the closest degree- the following settings were verified by hand
# and give the most accurate outputs

def closest_degree(degree):
    degree = degree.lower()
    b = ['bachelor', 'bs', 'ba']
    m = ['masters', 'mba', 'ms', 'meng', 'mcomp']
    p = ['dphil', 'phd', 'doc', 'post', 'grad']
    comb = ['bachelor', 'bs', 'ba', 'masters', 'mba', 'ms', 'dphil', 'phd', 'doc', 'post', 'grad']
    if difflib.get_close_matches(degree, b):
        return 'Bachelors'
    elif difflib.get_close_matches(degree, m):
        return 'Masters'
    elif difflib.get_close_matches(degree, p):
        return 'Doctorate'
    elif difflib.get_close_matches(degree, comb, 1, 0.4):
        out = difflib.get_close_matches(degree, comb, 1, 0.3)
        #print(degree, out)
        if out in b:
            return 'Bachelors'
        elif out in m:
            return 'Masters'
        elif out in p:
            return 'Doctorate'
        else:
            return 'Unknown'
    else:
        return 'Unknown'
```

```python
'''
The data set 'degrees of founders (unedited).xlsx' was labeled by hand.

It was created by all of the degrees that the founders obtained and we normalised it to be either a Doctorate degree or
above, masters, bachelors, associates, highschool or unknown. We assume all founders have completed highschool

Once we have normalised the diversity of dergree title to one of these few categories we then assign a value to each one 
Doctorate is 4
Masters is 2
Bachelors is 1
Associate is 0.5
others are 0

These numbers are absolutely arbitrary. 


One we have standardised the degrees of the founders, we the compute the best degree, worse degree, average degree score
'''

df_highest_education_map = pd.read_excel (r'degrees of founders (unedited).xlsx')
df_highest_education_map.drop(index = [0], inplace=True)
#print(df_highest_education_map)
standardised_degree_list = ['Doctorate', 'Masters', 'Unknown', 'Associate', 'Bachelors', 'DroppedOut', 'HS']
standardised_degree_list_and_values = [['Doctorate', 4],
                                       ['Masters',2],
                                       ['Unknown', 0],
                                       ['Associate', 0.5],
                                       ['Bachelors', 1],
                                       ['DroppedOut', 0],
                                       ['HS', 0]]

#Average Education Ranking

temp = df_All_Companies_merged['degrees_of_founders'].values[:].tolist()

df_All_Companies_merged['degrees_of_founders_standardised'] = float('nan')
df_All_Companies_merged['degrees_of_founders_standardised']  = df_All_Companies_merged['degrees_of_founders_standardised'] .astype(object)
df_All_Companies_merged['degrees_of_founders_standardised_numbers'] = float('nan')
df_All_Companies_merged['degrees_of_founders_standardised_numbers']  = df_All_Companies_merged['degrees_of_founders_standardised_numbers'] .astype(object)
df_All_Companies_merged['best_degree_of_founders'] = float('nan')
df_All_Companies_merged['worse_degree_of_founders'] = float('nan')
df_All_Companies_merged['average_degree_of_founders'] = float('nan')
df_All_Companies_merged['percentage_of_founders_with_a_degree'] = float('nan')

exceptional_degree_names = ['','N/A','NA']
flag_degree = set()

for idx in range(len(temp)):
    if not isinstance(temp[idx], str):
        if math.isnan(temp[idx]):
            continue
    temp_degree_names = temp[idx].split(',')
    number_of_degrees = len(temp_degree_names)
    temp_degree_standardised = []
    temp_degree_standardised_score = []
    for degree_name in temp_degree_names:
        if degree_name in exceptional_degree_names:
            temporary = 'Unknown'
        else:
            try:
                temporary = df_highest_education_map.loc[df_highest_education_map['degrees'] == degree_name, 'degrees_standardised'].values[0]
            except IndexError:
                flag_degree.add(degree_name)
                # if no match, find closest degree name with string simularity 
                temporary = closest_degree(degree_name)
                
        idx2 = standardised_degree_list.index(temporary)
        temp_degree_standardised.append(standardised_degree_list_and_values[idx2][0])
        temp_degree_standardised_score.append(standardised_degree_list_and_values[idx2][1])
    
    df_All_Companies_merged.at[idx, 'degrees_of_founders_standardised'] = ','.join([str(elem) for elem in temp_degree_standardised])
    df_All_Companies_merged.at[idx, 'degrees_of_founders_standardised_numbers'] = ','.join([str(elem) for elem in temp_degree_standardised_score])
    df_All_Companies_merged.at[idx, 'best_degree_of_founders'] = np.max(np.array(temp_degree_standardised_score))
    df_All_Companies_merged.at[idx, 'worse_degree_of_founders'] = np.min(np.array(temp_degree_standardised_score))
    df_All_Companies_merged.at[idx, 'average_degree_of_founders'] = np.average(np.array(temp_degree_standardised_score))
    df_All_Companies_merged.at[idx, 'percentage_of_founders_with_a_degree'] = number_of_degrees / df_All_Companies_merged.at[idx, 'number_of_founders']
#print(len(flag_degree))
#print(flag_degree)
df_All_Companies_merged[['degrees_of_founders_standardised','degrees_of_founders_standardised_numbers',
                         'best_degree_of_founders','worse_degree_of_founders',
                         'average_degree_of_founders','percentage_of_founders_with_a_degree']]
```

```python
'''
technical/stem degree 
1. flag is there is a stem subject
2. percentage of stem subjects in founders

'''



keywords = ['engineering', 'engineer', 'computer', 'science', 'mathematics', 'maths', 'physics', 'bio', 'chem', 'machine', 'statistics', 'medical', 'medicine','tech']

temp = df_All_Companies_merged['subject_degrees_of_founders'].values[:].tolist()

df_All_Companies_merged['stem_flag'] = float('nan')
df_All_Companies_merged['stem_percentage'] = float('nan')

for idx in range(len(temp)):
    if not isinstance(temp[idx], str):
        if math.isnan(temp[idx]):
            continue
    try: 
        temp_subject_names = temp[idx].split(',')
    except AttributeError:
        temp_subject_names = [str(temp[idx])]
    number_of_degrees = len(temp_subject_names)
    number_of_stem = 0
    for subject in temp_subject_names:
        if any(word in subject.lower() for word in keywords):
            number_of_stem += 1
            
    df_All_Companies_merged.at[idx, 'stem_percentage'] = number_of_stem / number_of_degrees
    if number_of_stem > 0:
        df_All_Companies_merged.at[idx, 'stem_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'stem_flag'] = 0
    continue

df_All_Companies_merged[['subject_degrees_of_founders','stem_flag', 'stem_percentage']]
```

```python
'''
Following flags: 
1. International University
2. Top International University 
'''

temp = df_All_Companies_merged['uni_country'].values[:].tolist()

df_All_Companies_merged['international_uni_flag'] = float('nan')
df_All_Companies_merged['international_uni_percentage'] = float('nan')

for idx in range(len(temp)):
    
    temp_country = temp[idx]

    number_of_countries = len(temp_country)
    number_of_int = 0
    for country in temp_country:
        if country != 'US':
            number_of_int += 1
            
    df_All_Companies_merged.at[idx, 'international_uni_percentage'] = number_of_int / number_of_countries
    if number_of_int > 0:
        df_All_Companies_merged.at[idx, 'international_uni_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'international_uni_flag'] = 0
    continue

df_All_Companies_merged[['uni_country','international_uni_flag', 'international_uni_percentage']]
```

```python
# Convert previous companies to list object
df_All_Companies_merged['prev_companies_of_founders'] = df_All_Companies_merged['prev_companies_of_founders'].replace(np.nan, '')
df_All_Companies_merged['prev_companies_of_founders'] = df_All_Companies_merged['prev_companies_of_founders'].apply(lambda x: list(x.split(',')) if type(x)==str else [])

# Convert previous companies to list object
prev_companies_of_founders = [item for sublist in df_All_Companies_merged['prev_companies_of_founders'].values for item in sublist]
prev_companies_of_founders = [x for x in prev_companies_of_founders if x!='']
prev_companies_of_founders_set = list(set(prev_companies_of_founders))

# Count the number of occurences of a unique company
prev_companies_of_founders_count = [[0, x] for x in prev_companies_of_founders_set]
for company in prev_companies_of_founders:
    idx = prev_companies_of_founders_set.index(company)
    prev_companies_of_founders_count[idx][0] += 1
prev_companies_of_founders_count.sort(reverse=True)

# Remove Inc., Inc and LLC
remove_list = [' Inc.', ' Inc', ' LLC']
for idx, value in enumerate(prev_companies_of_founders_count):
    if value[1] in remove_list:
        del prev_companies_of_founders_count[idx]

# Display the top ten most occurences
for idx, value in enumerate(prev_companies_of_founders_count[:10]):
    print('The top {} most popular previous company of founders is {} occuring {} times'.format(idx+1, value[1], value[0] ))
```

```python
'''
Prev Expereince 
1. big tech
2. unicorn 

'''


df_unicorn = pd.read_csv('unicorn.csv', delimiter=',')

list_of_unicorn = [list(row)[0] for row in df_unicorn.values]
list_of_big_tech = ['Google', 'Meta', 'Microsoft','Apple', 'Amazon', 'IBM', 'Tesla']

temp = df_All_Companies_merged['prev_companies_of_founders'].values[:].tolist()

df_All_Companies_merged['big_flag'] = float('nan')
df_All_Companies_merged['big_percentage'] = float('nan')
df_All_Companies_merged['unicorn_flag'] = float('nan')
df_All_Companies_merged['unicorn_percentage'] = float('nan')

for idx in range(len(temp)):
    
    
    temp_subject_names = temp[idx]
    
    number_of_comps = len(temp_subject_names)
    number_of_big = 0
    number_of_unicorn = 0
    
    for subject in temp_subject_names:  
        if subject in list_of_unicorn:
            number_of_unicorn += 1
        if subject in list_of_big_tech:
            number_of_big += 1
            
    df_All_Companies_merged.at[idx, 'big_percentage'] = number_of_big / number_of_comps
    if number_of_big > 0:
        df_All_Companies_merged.at[idx, 'big_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'big_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'unicorn_percentage'] = number_of_unicorn / number_of_comps
    if number_of_unicorn > 0:
        df_All_Companies_merged.at[idx, 'unicorn_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'unicorn_flag'] = 0


df_All_Companies_merged[['prev_companies_of_founders','unicorn_flag', 'big_flag']]
```

```python
'''
Prev Expereince in start ups (this is featured engineered as past experience in unicorns have high correlation)
1. start ups
2. unfunded
3. series c
4. series b
5. series a 
6. seed 
7. series c or unicorn 

'''

# Load the dataset provided
df_start_ups = pd.read_csv('Startups_per_stage_based_on_total_funding.csv', delimiter=',')

# Extract lists for each group
list_of_start_ups = [list(row)[0] for row in df_start_ups.values]
list_of_unicorn_start_ups = [list(row)[0] for row in df_start_ups.values if row[2] == 'series_c_or_unicorn']
list_of_seed_start_ups = [list(row)[0] for row in df_start_ups.values if row[2] == 'seed']
list_of_a_start_ups = [list(row)[0] for row in df_start_ups.values if row[2] == 'series_a']
list_of_b_start_ups = [list(row)[0] for row in df_start_ups.values if row[2] == 'series_b']
list_of_c_start_ups = [list(row)[0] for row in df_start_ups.values if row[2] == 'series_c']
list_of_unfunded_start_ups = [list(row)[0] for row in df_start_ups.values if row[2] == 'no_funding_reported']

temp = df_All_Companies_merged['prev_companies_of_founders'].values[:].tolist()

df_All_Companies_merged['start_up_flag'] = float('nan')
df_All_Companies_merged['c_unicorn_flag'] = float('nan')
df_All_Companies_merged['series_a_flag'] = float('nan')
df_All_Companies_merged['series_b_flag'] = float('nan')
df_All_Companies_merged['series_c_flag'] = float('nan')
df_All_Companies_merged['unfunded_flag'] = float('nan')
df_All_Companies_merged['seed_flag'] = float('nan')
df_All_Companies_merged['start_up_per'] = float('nan')
df_All_Companies_merged['c_unicorn_per'] = float('nan')
df_All_Companies_merged['series_a_per'] = float('nan')
df_All_Companies_merged['series_b_per'] = float('nan')
df_All_Companies_merged['series_c_per'] = float('nan')
df_All_Companies_merged['unfunded_per'] = float('nan')
df_All_Companies_merged['seed_per'] = float('nan')

for idx in range(len(temp)):
    temp_names = temp[idx]
    
    number_of_comps = len(temp_names)
    # number of each group (series a, b, c, seed, unfunded) as we count 
    n_s = 0
    n_u = 0
    n_a = 0
    n_b = 0
    n_c = 0
    n_seed = 0
    n_unfund = 0
    
    for subject in temp_names:  
        if subject in list_of_start_ups:
            n_s += 1
        if subject in list_of_unicorn_start_ups:
            n_u += 1
        if subject in list_of_seed_start_ups:
            n_seed += 1
        if subject in list_of_a_start_ups:
            n_a += 1
        if subject in list_of_b_start_ups:
            n_b += 1
        if subject in list_of_c_start_ups:
            n_c += 1
        if subject in list_of_unfunded_start_ups:
            n_unfund += 1
            
    df_All_Companies_merged.at[idx, 'start_up_per'] = n_s / number_of_comps
    if n_s > 0:
        df_All_Companies_merged.at[idx, 'start_up_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'start_up_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'c_unicorn_per'] = n_u / number_of_comps
    if n_u > 0:
        df_All_Companies_merged.at[idx, 'c_unicorn_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'c_unicorn_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'series_a_per'] = n_a / number_of_comps
    if n_a > 0:
        df_All_Companies_merged.at[idx, 'series_a_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'series_a_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'series_b_per'] = n_b / number_of_comps
    if n_b > 0:
        df_All_Companies_merged.at[idx, 'series_b_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'series_b_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'series_c_per'] = n_c / number_of_comps
    if n_c > 0:
        df_All_Companies_merged.at[idx, 'series_c_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'series_c_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'seed_per'] = n_seed / number_of_comps
    if n_seed > 0:
        df_All_Companies_merged.at[idx, 'seed_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'seed_flag'] = 0
        
    df_All_Companies_merged.at[idx, 'unfunded_per'] = n_unfund / number_of_comps
    if n_unfund > 0:
        df_All_Companies_merged.at[idx, 'unfunded_flag'] = 1
    else: 
        df_All_Companies_merged.at[idx, 'unfunded_flag'] = 0

df_All_Companies_merged[['prev_companies_of_founders','start_up_flag', 'series_a_flag']]
```
