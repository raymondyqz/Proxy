```python
#Import all the libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from functools import reduce
import math
import urllib
import difflib 
import seaborn as sns
```

```python
'''
Funtion to calculate the cor between features and success rate, present them in a box plot, and returns a ranked list 
of all the input featues and their cor. 

''' 
def generate_box_plots(corr_params, df_test, save_flag = False):
    feature_importance = []

    for param in corr_params:
        var1 = df_test[param][(np.isnan(df_test[param]) == False)] 
        var2 = df_test['success_flag'][(np.isnan(df_test[param]) == False)] 
        #print('The correlation between {} and {}: {}'.format('Success',param,var1.corr(var2)))
        feature_importance.append([param,var1.corr(var2)])
        
        plt.figure(figsize=(5, 5))
        sns.boxplot(x='success_flag',y=param,data=df_test, palette='rainbow')
        if save_flag:
            file_name = 'Figures\\' + param + '.png'
            plt.savefig(file_name)
        
    print('______________________________')
    sorted_features = sorted(feature_importance, key=lambda x: x[1], reverse=True)      
    for entry in sorted_features:
        print(entry) 
    return sorted_features
```

```python
# Load test data for success v fail and brand v fail

df_test_s = pd. read_pickle(r'Data/all_data_features_s.pkl')
df_test_b = pd. read_pickle(r'Data/all_data_features_b.pkl')
```

```python
# Load data for pre and post covid 

df_test_b_c = pd. read_pickle(r'Data/all_data_features_before_covid.pkl')
df_test_a_c = pd. read_pickle(r'Data/all_data_features_after_covid.pkl')
```

```python
# run for initial features 

corr_params = ['in_sf',
               'percentage_founders_in_sf',
               'US_uni_flag',
               'US_uni_percentage',
               'stanford_flag',
               'stanford_percentage',
               'big_flag',
               'big_percentage',
              'unicorn_flag',
              'unicorn_percentage',
              'number_of_founders',
              'percentage_of_male_founders',
              'number_of_male_founders',
              'number_of_female_founders',
              'maximum_founders_university_score',
               'minimum_founders_university_score',
               'average_founders_university_score',
               'maximum_founders_university_employer_score',
              'average_founders_university_employer_score',
              'best_degree_of_founders',
              'average_degree_of_founders',
              'percentage_of_founders_with_a_degree',
              'stem_flag',
              'stem_percentage']

generate_box_plots(corr_params, df_test_s, save_flag = True)
print('**************************')
print('**************************')
generate_box_plots(corr_params, df_test_b, save_flag = True)
```

```python
# run for all features - intial plus feature engineered new features  

corr_params = ['maximum_founders_university_score', 'minimum_founders_university_score', 'average_founders_university_score', 
               'average_founders_university_employer_score', 'maximum_founders_university_employer_score', 'US_uni_flag', 
               'US_uni_percentage', 'stanford_flag', 'stanford_percentage', 'MIT_flag', 'MIT_percentage', 'IVY_flag', 
               'IVY_percentage', 'top_int_flag', 'top_int_percentage', 'num_universities', 'number_of_founders', 
               'number_of_male_founders', 'number_of_female_founders', 'percentage_of_male_founders', 'in_sf', 'in_ny', 
               'in_miami', 'percentage_founders_in_sf', 'percentage_founders_in_ny', 'percentage_founders_in_miami', 
               'best_degree_of_founders', 'worse_degree_of_founders', 'average_degree_of_founders', 
               'percentage_of_founders_with_a_degree', 'post_covid_flag', 'stem_flag', 'stem_percentage', 
               'international_uni_flag', 'international_uni_percentage', 'big_flag', 'big_percentage', 
               'unicorn_flag', 'unicorn_percentage', 'start_up_flag', 'c_unicorn_flag', 'series_a_flag', 'series_b_flag', 
               'series_c_flag', 'unfunded_flag', 'seed_flag', 'start_up_per', 'c_unicorn_per', 'series_a_per', 
               'series_b_per', 'series_c_per', 'unfunded_per', 'seed_per']

success_importance = generate_box_plots(corr_params, df_test_s, save_flag = True)
print('**************************')
print('**************************')
brand_importance = generate_box_plots(corr_params, df_test_b, save_flag = True)
```

```python
# Plot on graph the relative importance(correlation) of the features 

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
features = [name[0] for name in success_importance]
data1 = [name[1] for name in success_importance]
data2 = [name[1] for name in brand_importance]
data = [data1, data2]
X = np.arange(len(features))
ax.bar(X + 0.00, data[0], color = 'g', width = 0.25, tick_label = 'Before')
ax.bar(X + 0.25, data[1], color = 'r', width = 0.25, tick_label = 'After')
plt.xticks([r + 0.25 for r in range(len(features))],
        [name[0] for name in success_importance], rotation = 90)
plt.title("Feature importance for success/brand investors")
plt.show()
```
